```python
print("Interval: [1,2]")
y = derivative.subs(x,1.5)
if y>0:
    print(y, ":Increasing")
if y<0:
    print(y, ":Decreasing")
```

```python
print("Interval: [2,3]")
y = derivative.subs(x,2.5)
if y>0:
    print(y, "Increasing")
if y<0:
    print(y, "Decreasing")
```

```python
# c) Evaluate function at endpoints
```

```python
y = (x**2)*exp(-x)
print("At x=1: ", y.subs(x, 1))
print("At x=3: ", y.subs(x, 3))
print("At critical value x=2: ", y.subs(x, 2))
```

```python
# d) Finding extrema of all those values
```

```python
max(y.subs(x,1), y.subs(x,2), y.subs(x,3))
```

```python
min(y.subs(x,1), y.subs(x,2), y.subs(x,3))
```

```python
### 6) Plot the parametric curve of x(t) = cos(t), y(t) = sin(t), z = t, on [0, 2pi]
```

```python
plot3d_parametric_line(cos(t), sin(t), t, (t, 0, 2*pi))
```

```python
### 7) Plot the parametric surface of x(u,v) = cos(u+v), y(u,v) = cos(u+v), z(u,v) = u-v, on [-2pi, 2pi] [-2pi, 2pi]
```

```python
u,v = symbols("u,v")
plot3d_parametric_surface(cos(u+v), cos(u+v), (u-v), (u, -2*pi, 2*pi), (v, -2*pi, 2*pi))
```
