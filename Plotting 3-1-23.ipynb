```python
x=np.linspace(-2,2,41)
y=np.exp(-x**2)*np.cos(2*np.pi*x)
plt.plot(x,y,alpha=0.4,label="Decaying Cosine", color="red", linestyle ="dashed", marker = "o", markersize = 5, 
         markerfacecolor = "blue")
plt.ylim(-2,2)
plt.legend()
plt.show()
```

```python
x = np.linspace(-np.pi, 2*np.pi, 100)
y = np.sin(x)
plt.plot(x,y,color="blue")
y1=x
plt.plot(x,y1,color="red")
y2=x-x**3/6
plt.plot(x,y2,color="green")
y3=x-x**3/6+x**5/120
plt.plot(x,y3,color="black")
```

```python
t=np.linspace(0,2*np.pi,100)
x=16*np.sin(t)**3
y=13*np.cos(t)-5*np.cos(2*t)-2*np.cos(3*t)-np.cos(4*t)
plt.plot(x,y,c=(1,0.2,0.5),lw=20)
plt.title("I LOVE MATH")
plt.axis("equal")
plt.axis("off")
plt.show()
```
